# Smart Phone Addiction Analysis

## Introduction
The rapid growth of smartphone usage has significantly transformed modern lifestyles, influencing how individuals communicate, work, and spend their leisure time. While smartphones offer convenience and connectivity, excessive usage has raised concerns regarding digital addiction and its impact on mental well-being, productivity, and daily habits.

This project utilizes a dataset comprising information from approximately 7,500 individuals, capturing detailed smartphone usage patterns and behavioral indicators. The dataset includes key variables such as daily screen time, number of app opens, time spent on social media, sleep duration, productivity levels, and frequency of phone checking. Additionally, demographic attributes like age, gender, and occupation enable a deeper understanding of how different user groups interact with smartphones.

The primary objective of this analysis is to explore patterns, trends, and relationships within the data to better understand smartphone usage behavior and its potential consequences. 

Through this analysis, the purpose is to identify the indicators of smartphone addiction, examine their impact on lifestyle factors such as sleep and productivity, and provide data-driven insights that can contribute to discussions around digital well-being.

### Questions and Hypotheses
1. **Does increased daily screen time affects an individual's productivity and effeciency?**
    - **Hypotheses**: Higher daily screen time is associated with lower productivity levels.


2. **Does spending more time on social media impact sleep duration?**
    - **Hypotheses**: Individuals who spend more time on social media tend to have shorter sleep durations.


3. **Do people spend more time on smartphones during weekends compared to weekdays?**
    - **Hypotheses**: Average screen time is higher on weekends than on weekdays


4. **Is there any relationship between daily screen time and stress_levels?**
    - **Hypotheses**: Often high duration of daily screen time is associated with high stress levels, but it is also possible that if properly used for entertainment it might decrease the stress


5. **Do individuals who spend more time gaming have different sleep durations?**
    - **Hypotheses**: Higher gaming hours are always associated with reduced sleep duration

## Step 3 - Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('Dataset/cleaned_data.csv')
df.head()
```

### 1. Does increased daily screen time affects an individual's productivity and effeciency?

a. Scatter Plot to know the productivity according to daily screen time 

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='daily_screen_time_hours', y='work_study_hours', palette='viridis')
plt.title('Daily Screen Time vs Productivity (Work/Study Hours)')
plt.xlabel('Daily Screen Time (hours)')
plt.ylabel('Work/Study Hours')
plt.show()
```

Since the scatter plot doesn't give any clear indication of impact of daily screen time, so it is possible that screen time doesn't affect productivity.

b. Lets check the trend of usage based on age factor, as it is possible that for different age groups the productivity has different impacts.

```python
# Checking for young people
young_people= df[df['age'] < 30]

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=young_people,
    x='daily_screen_time_hours',
    y='work_study_hours',
    hue='age'
)

plt.title('Screen Time vs Productivity - Grouped by Age')
plt.xlabel('Daily Screen Time in hours')
plt.ylabel('Work/Study Hours')

plt.show()
```

```python
# Checking for Older people group

old_people= df[df['age'] >= 30]

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=old_people,
    x='daily_screen_time_hours',
    y='work_study_hours',
    hue='age'
)

plt.title('Screen Time vs Productivity - Grouped by Age')
plt.xlabel('Daily Screen Time in hours')
plt.ylabel('Work/Study Hours')

plt.show()
```

Clearly age factor also doesn't impact the productivity/

c. Lets check the trend of usage based on gender, as it is possible that based on gender, the productivity has different impacts. 

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='daily_screen_time_hours',
    y='work_study_hours',
    hue='gender'
)
plt.title('Screen Time vs Productivity by Gender')
plt.xlabel('Daily Screen Time in Hours')
plt.ylabel('Work / Study Hours')
plt.show()
```

Clearly gender also doesn;t show difference in productivity

d. Since differnt age groups and gender clearly doesn't affect productivity, lets dig deep to check the impact after certain hours of screen. As it is possible that after certain hours productivity drops?

```python
low_screen_time = df[df['daily_screen_time_hours'] <= 5]
medium_screen_time = df[(df['daily_screen_time_hours'] > 5) & (df['daily_screen_time_hours'] <= 10)]
high_screen_time=df[df['daily_screen_time_hours'] > 10]
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=low_screen_time,
    x='daily_screen_time_hours',
    y='work_study_hours'
)

plt.title('Productivity vs Screen Time - Low Screen Time')
plt.xlabel('Daily Screen Time in Hours')
plt.ylabel('Work/Study Hours')

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=medium_screen_time,
    x='daily_screen_time_hours',
    y='work_study_hours'
)

plt.title('Productivity vs Screen Time - Medium Screen time')
plt.xlabel('Daily Screen Time in Hours')
plt.ylabel('Work/Study Hours')

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=high_screen_time,
    x='daily_screen_time_hours',
    y='work_study_hours'
)

plt.title('Productivity vs Screen Time - High Screen Time')
plt.xlabel('Daily Screen Time in Hours')
plt.ylabel('Work/Study Hours')

plt.show()
```

- All the above visualisations shows that there is no clear relationship between screen usage and productivity.
- Productivity might depend on the type of usage either work or entertainment, as some use screen for work and studies too.

e. Checking impact of social media usage to productivity

```python
df.head()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='social_media_hours',
    y='work_study_hours'
)

plt.title('Productivity vs Social Media Screen Time')
plt.xlabel('Social Media Hours')
plt.ylabel('Work/Study Hours')

plt.show()
```

### *Based on above visualisations, it is concluded that the results contradicts the hypotheses and hence there is no direct relationship of productivity to screen time.*

### 2. Does spending more time on social media impact sleep duration?


```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='social_media_hours',
    y='sleep_hours'
)

plt.title('Social Media Usage vs Sleep Duration')
plt.xlabel('Social Media Hours per Day')
plt.ylabel('Sleep Duration in hours')

plt.show()
```

Result: There is no direct relation of social media usage to sleeping time.

Lets check the affect due to gaming hours

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='gaming_hours',
    y='sleep_hours'
)

plt.title('Gaming Usage vs Sleep Duration')
plt.xlabel('Gaming Hours per Day')
plt.ylabel('Sleep Duration in hours')

plt.show()
```

### *Based on above visualisations, it is concluded that the results contradicts the hypotheses and hence there is no direct relationship of sleep time to either social media usage or gaming usage.*
